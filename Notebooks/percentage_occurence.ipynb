```python
bahamas = xr.open_dataset('/Users/dessyb/Box/Patrick/combined_netcdf/Bahamas_surfaceobs_1971-2020.nc')

#placing date variables into a separate dataframe
date_df = pd.DataFrame({'year':bahamas.year,'month':bahamas.month,'day':bahamas.day})

#creating datetime column
dates = pd.to_datetime(date_df)

dates

cl0 = []
cl1 = []
cl2 = []
cl3 = []
cl4 = []
cl5 = []
cl6 = []
cl7 = []
cl8 = []
cl9 = []

for i in range(len(bahamas.cl)):
    
    cl0.append(1 if bahamas.cl[i] == 0 else 0) 
    cl1.append(1 if bahamas.cl[i] == 1 else 0)
    cl2.append(1 if bahamas.cl[i] == 2 else 0)
    cl3.append(1 if bahamas.cl[i] == 3 else 0)
    cl4.append(1 if bahamas.cl[i] == 4 else 0)
    cl5.append(1 if bahamas.cl[i] == 5 else 0)
    cl6.append(1 if bahamas.cl[i] == 6 else 0)
    cl7.append(1 if bahamas.cl[i] == 7 else 0)
    cl8.append(1 if bahamas.cl[i] == 8 else 0)
    cl9.append(1 if bahamas.cl[i] == 9 else 0)

df1 = pd.DataFrame({'dates': dates,'cl0':cl0,'cl1':cl1,'cl2':cl2,'cl3':cl3,'cl4':cl4,'cl5':cl5,'cl6':cl6,'cl7':cl7,'cl8':cl8,'cl9':cl9})
df1 = df1.set_index('dates')
df1 = pd.DataFrame({'dates': dates,'cl0':cl0,'cl1':cl1,'cl2':cl2,'cl3':cl3,'cl4':cl4,'cl5':cl5,'cl6':cl6,'cl7':cl7,'cl8':cl8,'cl9':cl9})
df1 = df1.set_index('dates')
df2 = df1[df1.index.month < 4]

bahamas_perc_occ = df2.groupby(df2.index.year).mean()
```
